Exercise 1: FizzBuzz
Write a FizzBuzz Function: Create a function fizzbuzz(n) that takes an integer n as a parameter.
Implement FizzBuzz Logic: The function should print:
    - "Fizz" for multiples of 3
    - "Buzz" for multiples of 5
    - "FizzBuzz" for multiples of both 3 and 5
    - The number itself for other numbers
Call the Function: Call the function for numbers 1 to 20

```python
def fizzbuzz(n):
    for i in n:
        if i % 3 == 0 and i%5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else: 
            print(i)

fizzbuzz(range(1,21))
```

Exercise 2: Basic Data Filtering
- Create a List of Mixed Data Types: Create a list that contains a mix of integers, strings, and floats.
- Filter the List: Use list comprehension to create a new list that contains only the integers from the original list.
- Print the New List: Output the filtered list of integers.

```python
mix_list = ["Hello" , 1.2 , 1 , "Python" , 88.4 , 5 , 3 , True , True , False]

int_list = [ item for item in mix_list if type(item) == int]
print(int_list)
```

```python
#Extra answer where the user can input the type that he/she wants to filter
type_selected = input("Please enter the type you want to filter: ")
def type_selector(type_selected , user_list):
    if type_selected == "integer":
        int_list_2 = [ item for item in user_list if type(item) == int]
        print(int_list_2)
    elif type_selected == "string":
        str_list_2 = [ item for item in user_list if type(item) == str]
        print(str_list_2)
    elif type_selected == "float":
        float_list_2 = [ item for item in user_list if type(item) == float]
        print(float_list_2)
    elif type_selected == "bool":
        bool_list_2 = [ item for item in user_list if type(item) == bool]
        print(bool_list_2)
    else: 
        print("Your input is not a type. Please rewrite it")

type_selector(type_selected , mix_list)
```

Exercise 3: Simple To-Do List
1. Create an Empty List: Start with an empty list called todo_list.
2. Define Functions:
    - A function add_task(task) that adds a task to the list.
    - A function show_tasks() that prints all tasks in the list.

```python
todo = []

def add_task(task):
    todo.append(task)
    return todo

def show_tasks(todo):
    if todo:
        for task_item in todo:
            print(task_item)
    else: 
        print("To do list is currently empty.")

```

```python
#When the todo list is empty: 
show_tasks(todo)
```

```python
#When there are elements in the list:
add_task("Do the chores")
add_task("Finish Python Assignment")

show_tasks(todo)
```

Exercise 4: Temperature Converter
1. Define a Conversion Function: Write a function celsius_to_fahrenheit(celsius) that converts Celsius to Fahrenheit.
2. Print the Result: Output the converted temperature for 22ºF, 46ºF, 51ºF and 76ºF.

```python
def celsius_to_fahreinheit(celsius):
    fahrenheit = (9/5) * celsius + 32
    print(f"{celsius} celsius is equal to {fahrenheit} fahreinheit")

celsius_to_fahreinheit(22)
celsius_to_fahreinheit(46)
celsius_to_fahreinheit(51)
celsius_to_fahreinheit(76)
```
