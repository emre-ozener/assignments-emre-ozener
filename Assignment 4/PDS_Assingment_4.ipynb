#### Exercise 2:How many of them follow the name convention expressed above?


```python

counter = 0 #creating a counter to check the number of files
all_files= glob.glob('session_4/annotations/*.txt')


for file in all_files:
    file_name  = file.split('/')[-1]   # extract file name
    file_parts = file_name.split(".")[0].split("_") #remove the ".txt" part and split by "_"

    
    if len(file_parts) == 11: #check if the splitted part's length is 11 
        if file_parts[0].isdigit() and len(file_parts[0])==8: #if date part is all digits and 8 characters
            if file_parts[1].isdigit() and len(file_parts[1])==6: #if time part is all digits and 6 characters
                if file_parts[2].startswith("SN"): #if satellite starts with SN
                    if file_parts[3] == "QUICKVIEW" and file_parts[4] == "VISUAL": #if "QUICKVIEW" and "VISUAL" exists
                        if file_parts[5].isdigit() and file_parts[6].isdigit() and file_parts[7].isdigit():
                            if file_parts[8].startswith("SATL"):
                                if file_parts[9].isdigit() and file_parts[10].isdigit():
                                    counter = counter + 1 #counter increases as the conditions are fulfilled
print(counter)
```

#### Exercise 3: How many of annotations you have per month and year. Which month has more annotation files?


```python
months = {} #the dict for months
years = {} #the dict for years
for file in all_files:
    file_name  = file.split('/')[-1]   # Extract file name
    file_parts = file_name.split(".")[0].split("_") 
    file_date = file_parts[0]
    
    year = file_date[0:4] #determining the positions of the year
    month = file_date[4:6] #determining the positions of the month
    
    if year not in years.keys(): #if the year is not a key, it adds it 

        years[year] = 1
    else:
        years[year] += 1 #if yes, it increases the value by one
        
    if month not in months.keys(): #same procedure for months

        months[month] = 1
    else:
        months[month] += 1
print(years)
print(months)
```

#### Exercise 5: Print all the annotations from the most recent to the oldest one. 


```python
date_files = [] 

for file in all_files:
    file_names = os.path.basename(file)
    file_parts_2 = file_names.split('_')[0]
    file_with_date = [file_parts_2 , file_names]
    date_files.append(file_with_date)
print(date_files) #new nested lists that include date and the name of the file

```

```python
date_files = np.array(date_files) #make the list a np.array

date_files_sorted = np.sort(date_files , axis = 0)[::-1] #sort them according to the date and reverse it


for file in date_files_sorted:
    print(file[1]) # print the file name
```

#### Exercise 6: How many different satellites there are, how many annotations we have per satellite number, and which one was used in the most recent annotation file?


```python
counter = 0 
all_files= glob.glob('session_4/annotations/*.txt')
unique_satellites = {} #create a dict for satellites

for file in all_files:
    file_name  = file.split('/')[-1]   # extract file name
    file_parts = file_name.split(".")[0].split("_") 

    
    if len(file_parts) == 11: 
        if file_parts[0].isdigit() and len(file_parts[0])==8:
            if file_parts[1].isdigit() and len(file_parts[1])==6:
                if file_parts[2].startswith("SN"):
                    if file_parts[2] not in unique_satellites:#creates a key for each unique satellite
                        counter = counter + 1 #increases the counter by one
                        unique_satellites[file_parts[2]] = 1
                    else: unique_satellites[file_parts[2]] = unique_satellites[file_parts[2]] + 1 #increases the each satellite's number by one
print(f"There are {counter} unique satellites")
print(unique_satellites)
```

#### Exercise 7: How many unique regions are there?

```python
counter = 0 #setting the counter for regions
all_files= glob.glob('session_4/annotations/*.txt')
unique_regions = {} #dict for regions

for file in all_files:
    file_name  = file.split('/')[-1]   # Extract file name
    file_parts = file_name.split(".")[0].split("_") 

    
    if len(file_parts) == 11: 
        if file_parts[0].isdigit() and len(file_parts[0])==8:
            if file_parts[1].isdigit() and len(file_parts[1])==6:
                if file_parts[2].startswith("SN"):
                    if file_parts[3] == "QUICKVIEW" and file_parts[4] == "VISUAL":
                        if file_parts[5].isdigit() and file_parts[6].isdigit() and file_parts[7].isdigit():
                            if file_parts[8].startswith("SATL"):
                                if file_parts[9].isdigit() and file_parts[10].isdigit():
                                    file_parts_unique = file_parts[8] + file_parts[9] + file_parts[10] #combining the different parts of region
                                    if file_parts_unique not in unique_regions.keys():
                                        unique_regions[file_parts_unique] = 1 #adding the unique region to the dict
                                        counter = counter + 1 #increasing the counter by one each time there is a unique region
                                    else:
                                        unique_regions[file_parts_unique] = unique_regions[file_parts_unique] + 1 #increasing the dictionary value by one each time existing region comes
print(f"There are {counter} unique regions")
```
